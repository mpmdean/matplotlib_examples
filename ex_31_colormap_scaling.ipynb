```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from matplotlib.colors import LinearSegmentedColormap

%matplotlib inline
```

In matplotlib Color scales can straightforwardly be scaled in order to pick out changes towards the maximum or minimum of the range.

```python
power = 1/4

# Example data
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)
Z = np.abs(np.sin(X) * np.cos(Y))**2

fig, axs = plt.subplots(1, 2, figsize=(6+3/4, 3))

titles = ['Min emphasis', 'Max emphasis']
for ax, power, title in zip(axs, [1/4, 4], titles):
    cmap = LinearSegmentedColormap.from_list('Purples', plt.cm.Purples(np.linspace(0, 1, 100)**power))

    art = ax.pcolormesh(X, Y, Z, cmap=cmap, vmin=0, vmax=1)
    ax.set_aspect('equal')
    plt.colorbar(art, ax=ax, label='Value')
    ax.set_title(title)

    ax.xaxis.set_minor_locator(ticker.AutoMinorLocator(2))
    ax.yaxis.set_minor_locator(ticker.AutoMinorLocator(2))
              
fig.savefig('ex_31_colormap_scaling.pdf', dpi=1200)
```
