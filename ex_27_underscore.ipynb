```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

%matplotlib inline
```

Underscores can be created in python by using its capability to use unicode text. The character string \u0332 adds an underline to the preceeding character.

```python
fig, ax = plt.subplots(figsize=(3 + 3/8, 2.5))

ax.plot([0, 1], [0, 1], label='$d^8L\u0332$')
ax.set_xlabel('x-label')
ax.set_ylabel('y-label')
ax.set_title("Plot legend with underscore")
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator(2))
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator(2))

ax.legend()
fig.savefig('ex_27_underscore.pdf', dpi=1200)
```
