```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
plt.rcParams['figure.dpi'] = 1200
plt.rcParams['savefig.dpi'] = 1200


%matplotlib inline
```

```python
x = np.linspace(-6, 6, 1000)
y = 0.5*np.exp(-(x - 4)**2) + np.exp(-(x + 1)**2) + 0.3*np.exp(-(x + 4.5)**2)
# y = y/y.max()
```

```python
fig, ax = plt.subplots(figsize=(3 + 3/8, 2.5))

ax.plot(x, y, color=[.5]*3)

X, Y = np.meshgrid(x, y)
extent=(np.min(x), np.max(x), np.min(y), np.max(y))

ax.imshow(X, extent=extent, cmap=plt.get_cmap('Spectral'), aspect='auto')
ax.set_xlabel('Energy loss (eV)')
ax.set_ylabel('Intensity')

ax.fill_between(x, y, y.max(), color='w')
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator(2))
ax.set_yticks([])
plt.tight_layout()

fig.savefig('ex_25_rainbowspectrum.pdf', dpi=1200)
```
