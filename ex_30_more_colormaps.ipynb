```python
import matplotlib.pyplot as plt
import numpy as np
from cmcrameri import cm
import matplotlib.ticker as ticker

%matplotlib inline
```

Illustrating additional useful colormaps

```python
cm.show_cmaps()
```

```python
# Example plot
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

fig, ax = plt.subplots(figsize=(3+3/8, 3+3/8))

art = ax.contourf(X, Y, Z, levels=100, cmap=cm.davos)
ax.set_aspect('equal')
plt.colorbar(art, ax=ax, label='Value')
ax.set_title('Custom davos colormap')

ax.xaxis.set_minor_locator(ticker.AutoMinorLocator(2))
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator(2))

              
fig.savefig('ex_30_more_colormaps.pdf', dpi=1200)
```
